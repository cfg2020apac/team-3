```python
import requests as reqs
import pandas as pd
import json
```

```python
events_data = reqs.get('http://172.19.46.9:1337/events')
events_data = json.loads(events_data.text)
events_df = pd.DataFrame.from_records(events_data)
events_df.to_csv('sample-data/events.csv')

events_df.head()
```

```python
interests_data = reqs.get('http://172.19.46.9:1337/interests')
interests_data = json.loads(interests_data.text)
interests_df = pd.DataFrame.from_records(interests_data)
interests_df.to_csv('sample-data/interests.csv')

interests_df
```

```python
users_data = reqs.get('http://172.19.46.9:1337/volunteer-profiles')
users_data = json.loads(users_data.text)
users_data = pd.DataFrame.from_records(users_data)
users_data.to_csv('sample-data/users.csv')


users_data.head()
```
