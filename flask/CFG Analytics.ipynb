```python
users_data = pd.read_csv('sample-data/users.csv')
b15={}
ftt={}
a21={}
age_range={'below 15:':b15,'15-21':ftt,'above 21':a21}


types_of_events={}
for i in range(len(users_data)):
    s = users_data['interests'][i]
    json_acceptable_string = s.replace("'", "\"")
    d = json.loads(json_acceptable_string)
    ndf = pd.json_normalize(d)
    age = 2020 - int(users_data['birthdate'][i][:4])

    if age<15:
        for j in range(len(ndf)):
            if ndf['name'][j] in types_of_events.keys():
                b15[ndf['name'][j]]+=1
            else:
                b15[ndf['name'][j]]=1
    elif age<21:
        for j in range(len(ndf)):
            if ndf['name'][j] in types_of_events.keys():
                ftt[ndf['name'][j]]+=1
            else:
                ftt[ndf['name'][j]]=1
    else:
        for j in range(len(ndf)):
            if ndf['name'][j] in types_of_events.keys():
                a21[ndf['name'][j]]+=1
            else:
                a21[ndf['name'][j]]=1
print(b15)
print(ftt)
print(a21)
```
