```python
users_data = pd.read_csv('sample-data/users.csv')
types_of_events={}
for i in range(len(users_data)):
    s = users_data['interests'][i]
    json_acceptable_string = s.replace("'", "\"")
    d = json.loads(json_acceptable_string)
    ndf = pd.json_normalize(d)
    for j in range(len(ndf)):
        if ndf['name'][j] in types_of_events.keys():
            types_of_events[ndf['name'][j]]+=1
        else:
            types_of_events[ndf['name'][j]]=1
print(types_of_events)

```

```python
import matplotlib.pyplot as plt
plt.pie([float(v) for v in types_of_events.values()], labels=[str(k) for k in types_of_events.keys()], autopct=None)
plt.show()
```

```python
#count number of events 

for i in range(len(users_data)):
    s = users_data['events'][i]
    json_acceptable_string = s.replace("'", "\"")
    json_acceptable_string_1 = json_acceptable_string.replace('""', '","')
    final_str=json_acceptable_string_1[:-2]+json_acceptable_string_1[-1]
    print(final_str)
    d = json.loads(json_acceptable_string_1)
    ndf = pd.json_normalize(d)
    for j in range(len(ndf)):
        if ndf['name'][j] in types_of_events.keys():
            types_of_events[ndf['name'][j]]+=1
        else:
            types_of_events[ndf['name'][j]]=1
print(types_of_events)
```
